### Defaulter Rates per Housing Type

```python
a = application.groupby(['NAME_HOUSING_TYPE','TARGET']).count()[['SK_ID_CURR']]
a = (a/a.groupby(level=0).sum())*100
a
```

#### Clients with Rented Apartment and who live with Parents have a comparatively higher defaulter rates
#### at 12% and 11% respectively

```python
a = application.groupby(['NAME_HOUSING_TYPE','TARGET']).count()[['SK_ID_CURR']].reset_index()
plt.figure(figsize=(8,5))
sns.barplot(x='NAME_HOUSING_TYPE',y="SK_ID_CURR",hue="TARGET",data=a)
plt.ylabel("COUNT",size=12)
plt.xticks(rotation=45)
plt.show()
```

### Defaulter Rates per Income Type

```python
a = application.groupby(['NAME_INCOME_TYPE','TARGET']).count()[['SK_ID_CURR']]
(a/a.groupby(level=0).sum())*100
```

### Maternity and Unemployed clients have higher defaulter rates at 40% and 36% respectively
### Meanwhile Businesssman and Student clients have 100% Defaulter Rates 
### (10/10 and 18/18)

```python
a = application.groupby(['NAME_INCOME_TYPE','TARGET']).count()[['SK_ID_CURR']].reset_index()
plt.figure(figsize=(8,5))
sns.barplot(x='NAME_INCOME_TYPE',y="SK_ID_CURR",hue="TARGET",data=a)
plt.ylabel("COUNT",size=12)
plt.xticks(rotation=45)
plt.show()
```
