## Coeff Scores

```python
features = list(X_train.columns)
coeff = log.coef_
```

```python
coeff_score = pd.DataFrame(coeff.T,index=features,columns=['Coeff_score'])
```

```python
coeff_score.sort_values(by='Coeff_score',ascending=False,inplace=True)
```

```python
coeff_score.head()
```

```python
coeff_score.to_excel('coeff_scores.xlsx')
```

```python
def split_train(columns_df):
  X = columns_df
  y = df['TARGET']
  y = y.astype('int')

  X_train1,X_test,y_train1,y_test = train_test_split(X,y,test_size=0.3,random_state=0)

  X_train,X_val,y_train,y_val = train_test_split(X_train1,y_train1,test_size=0.2,random_state=0)
  #scores
  log = LogisticRegression(random_state=0)

  log.fit(X_train,y_train)

  y_pred = log.predict(X_val)

  print("Accuracy = {0} \nRecall {1} \nPrecision {2} \nF1 score {3}".format( 
        accuracy_score(y_val,y_pred),recall_score(y_val,y_pred),
        precision_score(y_val,y_pred),f1_score(y_val,y_pred)))
```

```python
abs_coeff_scores = np.abs(coeff_score)
print('Absolute Coeff Scores Sum = {0}'.format(abs_coeff_scores.sum()))
```

#### 100 columns

```python
print("Coeff Percentage = ",((abs_coeff_scores[:100].sum())/abs_coeff_scores.sum()*100)[0],"%")
imp_columns = list(abs_coeff_scores[:100].index)
imp_columns_df = df[imp_columns]
```

```python
split_train(imp_columns_df)
```
