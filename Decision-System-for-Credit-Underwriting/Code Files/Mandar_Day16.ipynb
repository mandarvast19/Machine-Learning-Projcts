## No of Previous Applications

```python
list_merged_df['No_of_PreviousApplications'] = list_merged_df['SK_ID_PREV_list'].str.len()
```

```python
((list_merged_df.iloc[:,122:].isna().sum())/len(list_merged_df)*100).sort_values(ascending=False).head(100)
```

```python
previous_bins = [0,5,10,15,20,60]
merged_df['Previous_bins'] = pd.qcut(merged_df['No_of_PreviousApplications'],q=5)
merged_df['Previous_bins'].value_counts()
```

```python
a = merged_df.groupby(['Previous_bins','TARGET']).count()[['SK_ID_CURR']]
a_level1 = a.groupby(level=0).sum()
a_default = a/a_level1*100
a['DEFAULTER_RATES'] = a_default
a = a.drop(0,level=-1).droplevel(level=-1).rename(columns={'SK_ID_CURR':'TOTALCOUNT'}).reset_index()
a['TOTALCOUNT'] = (a_level1/len(merged_df)*100).reset_index()['SK_ID_CURR']
a
```

```python
import matplotlib.ticker as mtick
fig, ax1 = plt.subplots(figsize=(10,6))


sns.pointplot(x='Previous_bins',y='TOTALCOUNT',data=a,ax=ax1,color='blue',legend=False)
ax1.yaxis.set_major_formatter(mtick.PercentFormatter(decimals=2))
ax1.set(ylabel='Total Clients in Percentages',ylim=(0,35))
ax2 = ax1.twinx()


sns.pointplot(x='Previous_bins',y='DEFAULTER_RATES',data=a,ax=ax2,color='orange',legend=False)
ax2.yaxis.set_major_formatter(mtick.PercentFormatter())
ax2.legend(labels=['Defaulter_rates'],bbox_to_anchor=(0.892,0.95))
ax1.legend(labels=['Total Clients in percentages'],bbox_to_anchor=(0.99,1))
ax2.set(ylim=(0,11))
ax2.grid(None)

plt.show()
```

## Source Score Average

```python
list_merged_df['EXT_SCORES_SUM'] = list_merged_df[['EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3']].sum(axis=1)
list_merged_df['EXT_SCORES_AVG'] = (list_merged_df[['EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3']].sum(axis=1))/3
list_merged_df['EXT_SCORES_AVG'].describe()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputed_df = list_merged_df.copy()
```

```python
imputer = SimpleImputer(strategy='mean')
imputed_df[['EXT_SOURCE_1','EXT_SOURCE_2']] = imputer.fit_transform(imputed_df[['EXT_SOURCE_1','EXT_SOURCE_2']])
```

```python
imputed_df[['EXT_SOURCE_1','EXT_SOURCE_2']]
```

```python
imputed_df['EXT_SOURCE_1'].corr(imputed_df['TARGET'])
```

```python
imputed_df['EXT_SCORES_SUM'] = imputed_df[['EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3']].sum(axis=1)
imputed_df['EXT_SCORES_AVG'] = (imputed_df[['EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3']].sum(axis=1))/3
imputed_df['EXT_SCORES_AVG'].describe()
```

```python
import matplotlib.ticker as mtick
fig, ax1 = plt.subplots(figsize=(15,6))


sns.pointplot(x='SCORES_AVG_BINS',y='TOTALCOUNT',data=a,ax=ax1,color='blue',legend=False)
ax1.yaxis.set_major_formatter(mtick.PercentFormatter(decimals=2))
ax1.set(ylabel='Total Clients in Percentages',ylim=(20,30))
ax2 = ax1.twinx()


sns.pointplot(x='SCORES_AVG_BINS',y='DEFAULTER_RATES',data=a,ax=ax2,color='orange',legend=False)
ax2.yaxis.set_major_formatter(mtick.PercentFormatter())
ax2.legend(labels=['Defaulter_rates'],bbox_to_anchor=(0.9235,0.95))
ax1.legend(labels=['Total Clients in percentages'],bbox_to_anchor=(0.99,1))
ax2.set(ylim=(0,17))
ax1.set_xticklabels(a['SCORES_AVG_BINS'],rotation=45)
ax2.grid(None)

plt.show()
```
